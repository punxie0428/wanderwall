```python
# 필요한 라이브러리 임포트
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# 데이터 로드
data = pd.read_csv('C:/Down/penguins.csv') 
```

```python
# 데이터 확인
print(data.head())
```

```python
Image("C:/Down/0001.png")
```

```python
# 결측치 처리
data = data.dropna()
```

```python
# 'year' 열 드롭
data = data.drop(columns=['year'])
```

```python
# 'year'열 사라졌나 데이터 확인
print(data.head())
```

```python
# LabelEncoder 인스턴스 생성
label_encoder_species = LabelEncoder()
label_encoder_sex = LabelEncoder()
```

```python
# species와 sex 인코딩
data['species'] = label_encoder_species.fit_transform(data['species'])
data['sex'] = label_encoder_sex.fit_transform(data['sex'].astype(str))
```

```python
# 특성과 레이블 분리
X = data[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g', 'sex']]
y = data['species']
```

```python
# 데이터 분할 (훈련 데이터와 테스트 데이터)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=55)
```

```python
# 랜덤 포레스트 모델 생성 및 훈련
model = RandomForestClassifier(n_estimators=100, random_state=55)
model.fit(X_train, y_train)
```

```python
# 예측 수행
y_pred = model.predict(X_test)
```

```python
# 평가 결과 출력
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# 인코딩된 값의 종 확인
species_mapping = label_encoder_species.inverse_transform([0, 1, 2])  # 인코딩된 값으로 대체
print("인코딩된 값의 종:", species_mapping)
```
